# Frequency Counter

```python
arr = [1,2,1,3,2,1,4]

freq = {}

for num in arr:
    freq[num] = freq.get(num, 0) + 1

print(freq)
```

# First Non-Repeating Character

```python
s = "programming"

freq = {}

for i in s:
    freq[i] = freq.get(i,0)+1

for i in s:
    if freq[i] == 1:
        print(i)
        break
```

# enumerate()
- enumerate() is a built-in Python function that adds a counter (index) to an iterable (list, tuple, string, etc.) and returns an enumerate object.
- Instead of manually maintaining an index variable, enumerate() gives you both the index and the value while iterating

# Syntex
- enumerate(iterable, start=0)

| Parameter  | Description                        |
| ---------- | ---------------------------------- |
| `iterable` | List, tuple, string, etc.          |
| `start`    | Starting index value (default = 0) |


```python
fruits = ["Apple", "Banana", "Mango"]
```

```python
for i in enumerate(fruits):
    print(i)
```

```python
fruits = ["Apple", "Banana", "Mango"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
```

```python
name = ["Munna","Raunak","Suman","Manish"]
for index,value in enumerate(name,start=1):
    print(value)
```

```python
name = ["Munna","Raunak","Suman","Manish"]
for index,value in enumerate(name,start=1):
    print(index,value)
```

# zip() in Python
- zip() combines two or more iterables element-wise and returns a zip object.
- Think of it as creating pairs.
- Returns a zip object.



```python
print(type(zip([1], [2])))
```

```python
names = ["Rahul", "Aman", "John"]

marks = [90, 85, 95]

result = zip(names, marks)
```

```python
result
```

```python
list(result)
```

```python
names = ["Rahul", "Aman", "John"]

marks = [90, 85, 95]

for a,b in zip(names,marks):
    print(a,b)
```

```python
names = ["Rahul", "Aman"]

marks = [90, 85]

cities = ["Delhi", "Mumbai"]

for name, mark, city in zip(names, marks, cities):
    print(name, mark, city)
```

# Unzip a Zipped Object

```python
pairs = [
    ('Rahul', 90),
    ('Aman', 85)
]

names, marks = zip(*pairs)

print(names)
print(marks)
```

| Feature    | enumerate()              | zip()                                |
| ---------- | ------------------------ | ------------------------------------ |
| Purpose    | Add index                | Combine iterables                    |
| Input      | One iterable             | Multiple iterables                   |
| Returns    | `(index, value)` tuples  | Combined tuples                      |
| Common Use | Need index while looping | Loop through multiple lists together |


| Function                 | Returns                                    | Main Use                            |
| ------------------------ | ------------------------------------------ | ----------------------------------- |
| `enumerate(iterable)`    | Enumerate object → `(index, value)` tuples | Get index and value together        |
| `zip(iter1, iter2, ...)` | Zip object → Combined tuples               | Iterate multiple iterables together |
