# Lists vs Arrays

| Feature              | List                                                               | Array                                               |
| -------------------- | ------------------------------------------------------------------ | --------------------------------------------------- |
| **Data Types**       | Can store different data types (e.g., integers, strings, booleans) | Usually stores elements of the same data type       |
| **Size**             | Dynamic (can grow or shrink easily)                                | Generally fixed-size or less flexible               |
| **Performance**      | Slower for numerical computations                                  | Faster and more memory-efficient for numerical data |
| **Memory Usage**     | Uses more memory                                                   | Uses less memory                                    |
| **Built-in Support** | Built into Python                                                  | Requires modules like `array` or `NumPy`            |
| **Use Case**         | General-purpose data storage                                       | Mathematical and scientific computations            |

# Accessing Items in a List

# 1. Accessing Items Using Positive Indexing

Indexing starts from 0

```python
ruits = ["apple", "banana", "mango"]

print(fruits[0])  # apple
print(fruits[1])  # banana
print(fruits[2])  # mango
```

# 2. Accessing Items Using Negative Indexing

Negative indexing starts from -1 and moves backward.

```python
fruits = ["apple", "banana", "mango"]

print(fruits[-1])  # mango
print(fruits[-2])  # banana
```

# 3. Accessing Multiple Items Using Slicing

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])
```

# 4. Accessing Items in a Nested List

```python
matrix = [[1, 2], [3, 4], [5, 6]]

print(matrix[1][0])
```

# Editing Items in a List

Since lists are mutable, their elements can be modified after creation.

# 1. Editing a Single Item

```python
fruits = ["apple", "banana", "mango"]

fruits[1] = "orange"

print(fruits)
```

# 2. Editing Multiple Items Using Slicing

```python
numbers = [10, 20, 30, 40, 50]

numbers[1:4] = [25, 35, 45]

```

```python
numbers
```

# 3. Editing Items in a Nested List

```python
matrix = [[1, 2], [3, 4], [5, 6]]

matrix[1][1] = 40

print(matrix)
```

# Operations on Lists

Python provides several operators that can be used to perform operations on lists.

# 1. Concatenation (+)

The + operator is used to combine two or more lists into a single list.

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = list1 + list2
```

```python
result
```

# 2. Repetition (*)

The * operator is used to repeat the elements of a list.

```python
numbers = [1, 2, 3]

result = numbers * 3
```

```python
result
```

#  3. Membership Operators (in and not in)

These operators are used to check whether an item exists in a list.

```python
fruits = ["apple", "banana", "mango"]
```

```python
"banana" in fruits
```

```python
"banasssna" in fruits
```

# 4. Iteration

Lists can be traversed using loops.

```python
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
```

# 5. Comparison Operators

Lists can be compared using comparison operators.

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = [1, 2, 4]

print(list1 == list2)
print(list1 != list3)
```

# 6. Length Operation (len())

Returns the number of items in a list.

```python
fruits = ["apple", "banana", "mango"]

print(len(fruits))
```

# 7. List Slicing

Extract a portion of a list.

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])
```

# 8. List Traversal with Index

Access both index and value using enumerate().

```python
fruits = ["apple", "banana", "mango"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
```

| Operation            | Operator/Function    | Example             |
| -------------------- | -------------------- | ------------------- |
| Concatenation        | `+`                  | `list1 + list2`     |
| Repetition           | `*`                  | `list * 3`          |
| Membership           | `in`, `not in`       | `"apple" in fruits` |
| Iteration            | `for` loop           | `for item in list`  |
| Comparison           | `==`, `!=`, `<`, `>` | `list1 == list2`    |
| Length               | `len()`              | `len(list)`         |
| Slicing              | `[:]`                | `list[1:4]`         |
| Traversal with Index | `enumerate()`        | `enumerate(list)`   |


# Functions and Methods of Lists

Python provides several built-in functions and list methods to perform common operations on lists.

# 1. len()

Returns the number of items in a list.

```python
numbers = [10, 20, 30, 40]

print(len(numbers))
```

# 2. max()

Returns the largest item in a list.

```python
numbers = [10, 20, 30, 40]

print(max(numbers))
```

# 3. min()

Returns the smallest item in a list

```python
numbers = [10, 20, 30, 40]

print(min(numbers))
```

# 4. sum()

Returns the sum of all numeric items.

```python
numbers = [10, 20, 30, 40]

print(sum(numbers))
```

# 5. sorted()

Returns a new sorted list.

```python
numbers = [40, 10, 30, 20]

print(sorted(numbers))
```

# List Methods

List methods are functions that belong specifically to list objects.

# 1. append()

Adds an item to the end of the list.

```python
fruits = ["apple", "banana"]

fruits.append("mango")

print(fruits)
```

# 2. extend()

Adds all items from another iterable to the list.

```python
numbers = [1, 2, 3]

numbers.extend([4, 5, 6])

print(numbers)
```

# 3. insert()

Inserts an item at a specified position.

```python
fruits = ["apple", "mango"]

fruits.insert(1, "banana")

print(fruits)
```

# 4. remove()

Removes the first occurrence of a specified value

```python
fruits = ["apple", "banana", "mango"]

fruits.remove("banana")

print(fruits)
```

# 5. pop()

Removes and returns an item at a given index.

```python
fruits = ["apple", "banana", "mango"]

item = fruits.pop()

print(item)
print(fruits)
```

# 6. clear()

Removes all items from the list

```python
fruits = ["apple", "banana", "mango"]

fruits.clear()

print(fruits)
```

# 7. index()

Returns the index of the first occurrence of a value.

```python
fruits = ["apple", "banana", "mango"]

print(fruits.index("banana"))
```

# 8. count()

Returns the number of occurrences of a value.

```python
numbers = [1, 2, 2, 3, 2]

print(numbers.count(2))
```

# 9. sort()

Sorts the list in ascending order.

```python
numbers = [40, 10, 30, 20]

numbers.sort()

print(numbers)
```

# 10. reverse()

Reverses the order of items in the list.

```python
numbers = [1, 2, 3, 4]

numbers.reverse()

print(numbers)
```

# 11. copy()

Returns a shallow copy of the list.

```python
fruits = ["apple", "banana", "mango"]

new_fruits = fruits.copy()

print(new_fruits)
```
