# Set
- Unordered collection of unique elements.

```python
fruits = {"Apple", "Mango", "Banana"}
```

```python
fruits
```

# Duplicate Values Not Allowed 

```python
fruits = {"Apple", "Mango", "Banana", "Apple"}
```

```python
fruits
```

# Key Characteristics
- Stores unique values only (no duplicates)
- Unordered (items don't have fixed positions)
- Mutable (you can add or remove items)
- Faster lookups compared to lists

# Adding Elements

# Adding Multiple Elements

```python
friends = {"Sonu"}

friends.update(["Raju", "Monu", "Chintu"])
```

# Removing Elements
- remove()

```python
friends = {"Sonu", "Raju", "Monu"}

friends.remove("Raju")
```

```python
friends
```

# discard()

Safer version of remove.

```python
friends.discard("Aman")
```

# Checking Membership

```python
friends = {"Sonu", "Raju", "Monu"}

print("Sonu" in friends)
print("Aman" in friends)
```

# Looping Through a Set

```python
friends = {"Sonu", "Raju", "Monu"}

for friend in friends:
    print(friend)
```

# Set Operations

# 1. Union

Combines all unique elements.

```python
munna = {"Sonu", "Raju"}
```

```python
munna
```

```python
chintu = {"Monu", "Raju"}
```

```python
chintu
```

```python
all_friends = munna.union(chintu)
```

```python
all_friends
```

```python
# "Raju" appears only once.
```

# 2. Intersection

Finds common elements.

```python
munna = {"Sonu", "Raju", "Monu"}
chintu = {"Monu", "Raju", "Aman"}

common = munna.intersection(chintu)
```

```python
common
```

# 3. Difference

Elements present in first set but not second.

```python
munna = {"Sonu", "Raju", "Monu"}
chintu = {"Monu", "Aman"}

munna.difference(chintu)
```

# 4. Symmetric Difference

Elements that are not common.

```python
munna = {"Sonu", "Raju", "Monu"}
chintu = {"Monu", "Aman"}
```

```python
munna.symmetric_difference(chintu)
```

| Method                   | Purpose                       |
| ------------------------ | ----------------------------- |
| `add()`                  | Add one item                  |
| `update()`               | Add multiple items            |
| `remove()`               | Remove item (error if absent) |
| `discard()`              | Remove item safely            |
| `pop()`                  | Remove random item            |
| `clear()`                | Remove all items              |
| `copy()`                 | Create a copy                 |
| `union()`                | Combine sets                  |
| `intersection()`         | Common items                  |
| `difference()`           | Unique items                  |
| `symmetric_difference()` | Non-common items              |


| Method                   | Time Complexity        | Purpose               |
| ------------------------ | ---------------------- | --------------------- |
| `add()`                  | O(1)                   | Add one element       |
| `update()`               | O(n)                   | Add multiple elements |
| `remove()`               | O(1)                   | Remove element        |
| `discard()`              | O(1)                   | Remove safely         |
| `pop()`                  | O(1)                   | Remove random element |
| `clear()`                | O(n)                   | Empty set             |
| `copy()`                 | O(n)                   | Create copy           |
| `union()`                | O(len(A)+len(B))       | Combine sets          |
| `intersection()`         | O(min(len(A), len(B))) | Common elements       |
| `difference()`           | O(len(A))              | Unique elements       |
| `symmetric_difference()` | O(len(A)+len(B))       | Non-common elements   |
| `issubset()`             | O(len(A))              | Subset check          |
| `issuperset()`           | O(len(B))              | Superset check        |
| `isdisjoint()`           | O(min(len(A), len(B))) | Check common elements |


| Operation            | Symbol | Purpose                            | Python                      | Result                                        |
| -------------------- | ------ | ---------------------------------- | --------------------------- | --------------------------------------------- |
| Union                | ∪      | All unique elements from both sets | `A.union(B)`                | `{'Munna', 'Sonu', 'Raju', 'Monu', 'Chintu'}` |
| Intersection         | ∩      | Common elements in both sets       | `A.intersection(B)`         | `{'Sonu'}`                                    |
| Difference           | −      | Elements in A but not in B         | `A.difference(B)`           | `{'Munna', 'Raju'}`                           |
| Difference           | −      | Elements in B but not in A         | `B.difference(A)`           | `{'Monu', 'Chintu'}`                          |
| Symmetric Difference | △      | Elements present in only one set   | `A.symmetric_difference(B)` | `{'Munna', 'Raju', 'Monu', 'Chintu'}`         |
