# Pratice Problem based on the TUPLE

01. Create a tuple containing: Munna, 24, CSE, 2024, 8.98, AI Engineer, Data Science, Cricketer.
02. Print the entire tuple.
03. Print only the name from the tuple using indexing.
04. Print only the age from the tuple.
05. Print only the branch (CSE) from the tuple.
06. Print only the CGPA from the tuple.
07. Print the last element of the tuple using negative indexing.
08. Print the second-last element using negative indexing.
09. Print the first four elements using slicing.
10. Print the last three elements using slicing.
11. Find the total number of elements in the tuple.
12. Check whether "Munna" exists in the tuple.
13. Check whether "Python Developer" exists in the tuple.
14. Use a for loop to print all elements of the tuple.
15. Use a while loop to print all elements of the tuple.
16. Unpack the tuple into variables: name, age, branch, year, cgpa, profession, specialization, hobby.
17. Print only name, profession, and hobby after unpacking.
18. Create a new tuple by adding "Punjab" to the existing tuple.
19. Count how many times "CSE" appears in the tuple.
20. Find the index position of "AI Engineer" in the tuple.

# Solution of above problem and EXTRA 

```python
# 1. Create a tuple
student = ("Munna", 24, "CSE", 2024, 8.98, "AI Engineer", "Data Science", "Cricketer")

# 2. Print the entire tuple
print(student)

# 3. Print only the name
print(student[0])

# 4. Print only the age
print(student[1])

# 5. Print only the branch
print(student[2])

# 6. Print only the CGPA
print(student[4])

# 7. Print the last element
print(student[-1])

# 8. Print the second-last element
print(student[-2])

# 9. Print the first four elements
print(student[:4])

# 10. Print the last three elements
print(student[-3:])

# 11. Find total number of elements
print(len(student))

# 12. Check whether "Munna" exists
print("Munna" in student)

# 13. Check whether "Python Developer" exists
print("Python Developer" in student)

# 14. Use a for loop to print all elements
for item in student:
    print(item)

# 15. Use a while loop to print all elements
i = 0
while i < len(student):
    print(student[i])
    i += 1

# 16. Unpack the tuple
name, age, branch, year, cgpa, profession, specialization, hobby = student

# 17. Print name, profession, and hobby
print(name)
print(profession)
print(hobby)

# 18. Create a new tuple by adding "Punjab"
new_student = student + ("Punjab",)
print(new_student)

# 19. Count how many times "CSE" appears
print(student.count("CSE"))

# 20. Find index position of "AI Engineer"
print(student.index("AI Engineer"))

# 21. Reverse the tuple using slicing
print(student[::-1])

# 22. Print every second element
print(student[::2])

# 23. Print elements from index 2 to 5
print(student[2:6])

# 24. Create a tuple containing only numeric values
numeric_values = (student[1], student[3], student[4])
print(numeric_values)

# 25. Find the maximum numeric value
print(max(numeric_values))

# 26. Find the minimum numeric value
print(min(numeric_values))

# 27. Convert tuple into list
student_list = list(student)
print(student_list)

# 28. Convert list back into tuple
student_tuple = tuple(student_list)
print(student_tuple)

# 29. Concatenate another tuple
extra = ("Python", "ML")
combined = student + extra
print(combined)

# 30. Repeat tuple 2 times
print(student * 2)

# 31. Check whether "Data Science" is present
print("Data Science" in student)

# 32. Find index of "Cricketer"
print(student.index("Cricketer"))

# 33. Print tuple without the first element
print(student[1:])
```

# Find Frequency of Every Element in a List

```python
numbers = [1,2,3,1,2,1,4,5,2]
```

```python
for item in numbers:
    print(item, "->", numbers.count(item))
```

```python
numbers = [1,2,3,1,2,1,4,5,2]

visited = []

for item in numbers:
    if item not in visited:
        print(item, "->", numbers.count(item))
        visited.append(item)
```
